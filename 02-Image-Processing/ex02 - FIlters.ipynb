## Use of filters 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Original image**

```python
img = cv2.imread('../DATA/noise.png', 1)
plt.imshow(img)
```

**Average**

```python
# average filtering (replace pixels under kernel with their average)
img_avg = cv2.blur(img, (17,17))
plt.imshow(img_avg)
```

**Median**

```python
# median filtering (replace pixels under kernel with their median)
img_med = cv2.medianBlur(img, 17)
plt.imshow(img_med)
```

**Gaussian**

```python
# gaussian blur (gaussian kernel), effective in removing Gaussian noise from the image (kernel size MUST BE ODD!!!)
img_gauss = cv2.GaussianBlur(img, ksize=(17,17), sigmaX=0)
plt.imshow(img_gauss)
```

**Bilinear**

```python
# bilateral filter
img_bil = cv2.bilateralFilter(img, 17, sigmaSpace = 75, sigmaColor =75)
plt.imshow(img_bil)
```
