# Colorspaces

Let's have a brief introduction into converting to different colorspaces! The video goes into more detail about colorspaces.

Quick Info Link: 
https://en.wikipedia.org/wiki/HSL_and_HSV


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

OpenCV reads images in RGB!!!
Matplotlib displays them in BGR, so if I just show the image resulting colors will be messed up! 

```python
img = cv2.imread('../DATA/00-puppy.jpg')
plt.imshow(img)
```

An alternative is to show image by using cv2.imshow() method from OpenCV (it accepts the image natively read in BGR by cv2.imread())

```python
cv2.imshow('window', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Converting to Different Colorspaces

If I convert image from BGR to RGB I can display it with matplotlib

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

**Converting to HSV**
https://en.wikipedia.org/wiki/HSL_and_HSV

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

```python
img = cv2.imread('../DATA/00-puppy.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
